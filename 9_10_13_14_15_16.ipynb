```python
'''
Assignment - 9: Write a program to do the following: You have given a collection of 8 points. P1=[0.1,0.6] P2=[0.15,0.71] P3=[0.08,0.9] P4=[0.16, 0.85] P5=[0.2,0.3] P6=[0.25,0.5] P7=[0.24,0.1] P8=[0.3,0.2]. Perform the k-mean clustering with initial centroids as m1=P1=Cluster#1=C1 and m2=P8=cluster#2=C2. Answer the following 1] Which cluster does P6 belong to? 2] What is the population of a cluster around m2? 3] What is
the updated value of m1 and m2?
'''
```

```python
# Importing Libraries

import math
import pandas as pd
import matplotlib.pyplot as plt

# Inserting 8 Points
dataset = {'Points': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8'],
           'X': [0.1, 0.15, 0.08, 0.16, 0.2, 0.25, 0.24, 0.3],
           'Y': [0.6, 0.71, 0.9, 0.85, 0.3, 0.5, 0.1, 0.2]}

df = pd.DataFrame(dataset)
```

```python
x = df.X
y = df.Y
plt.scatter(x, y)
for i, txt in enumerate(df.Points):
    plt.annotate(txt, (df.X[i], df.Y[i]))

plt.show()
```

```python
def calculate_distance(p1, p2):
    x1 = p1[0]
    y1 = p1[1]
    x2 = p2[0]
    y2 = p2[1]
    return math.sqrt(math.pow((x2 - x1), 2) + math.pow((y2 - y1), 2))
```

```python
def update(m, p):
    x1 = m[0]
    y1 = m[1]
    x2 = p[0]
    y2 = p[1]
    return [(x1 + x2) / 2, (y1 + y2) / 2]
```

```python
points = ["P" + str(x + 1) for x in range(8)]
```

```python
coordinates = [[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85], [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]]
```

```python
m1 = coordinates[0]  # Cluster 1
m2 = coordinates[7]  # Cluster 2
m1_list = []
m2_list = []
```

```python
for i in range(8):
    print(f"Point Under Consideration {points[i]} : {coordinates[i]}")
    dis_m1 = calculate_distance(m1, coordinates[i])
    dis_m2 = calculate_distance(m2, coordinates[i])
    if dis_m1 < dis_m2:
        print(f"Point {points[i]} : {coordinates[i]} Added to Cluster 1")
        m1 = update(m1, coordinates[i])
        m1_list.append(points[i])
    else:
        print(f"Point {points[i]} : {coordinates[i]} Added to Cluster 2")
        m2 = update(m2, coordinates[i])
        m2_list.append(points[i])

print(f"\nPoints in Cluster 1 : {len(m1_list)}")
print(m1_list)
print(f"Points in Cluster 2 : {len(m2_list)}")
print(m2_list)

print(f"\nFinal value of m1 = {m1}")
print(f"Final value of m2 = {m2}")
```

```python
# Importing Libraries

import math
import pandas as pd
import matplotlib.pyplot as plt

# Inserting 8 Points
dataset = {'Points': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8'],
           'X': [2, 2, 8, 5, 7, 6, 1, 4],
           'Y': [10, 5, 4, 8, 5, 4, 2, 9]
        }

df = pd.DataFrame(dataset)
```

```python
x = df.X
y = df.Y
plt.scatter(x, y)
for i, txt in enumerate(df.Points):
    plt.annotate(txt, (df.X[i], df.Y[i]))

plt.show()
```

```python
def calculate_distance(p1, p2):
    x1 = p1[0]
    y1 = p1[1]
    x2 = p2[0]
    y2 = p2[1]
    return math.sqrt(math.pow((x2 - x1), 2) + math.pow((y2 - y1), 2))
```

```python
def update(m, p):
    x1 = m[0]
    y1 = m[1]
    x2 = p[0]
    y2 = p[1]
    return [(x1 + x2) / 2, (y1 + y2) / 2]
```

```python
points = ["P" + str(x + 1) for x in range(8)]
```

```python
coordinates = [[2, 10], [2, 5], [8, 4], [5, 8], [7, 5], [6, 4], [1, 2], [4, 9]]
```

```python
m1 = coordinates[0]  # Cluster 1
m2 = coordinates[3]  # Cluster 2
m3 = coordinates[6]  # Cluster 3
m1_list = []
m2_list = []
m3_list = []
```

```python
for i in range(8):
    print(f"Point Under Consideration {points[i]} : {coordinates[i]}")
    dis_m1 = calculate_distance(m1, coordinates[i])
    dis_m2 = calculate_distance(m2, coordinates[i])
    dis_m3 = calculate_distance(m3, coordinates[i])
    if dis_m1 == min(dis_m1, dis_m2, dis_m3):
        print(f"Point {points[i]} : {coordinates[i]} Added to Cluster 1")
        m1 = update(m1, coordinates[i])
        m1_list.append(points[i])
    elif dis_m2 == min(dis_m1, dis_m2, dis_m3):
        print(f"Point {points[i]} : {coordinates[i]} Added to Cluster 2")
        m2 = update(m2, coordinates[i])
        m2_list.append(points[i])
    else:
        print(f"Point {points[i]} : {coordinates[i]} Added to Cluster 3")
        m3 = update(m3, coordinates[i])
        m3_list.append(points[i])

print(f"\nPoints in Cluster 1 : {len(m1_list)}")
print(m1_list)
print(f"Points in Cluster 2 : {len(m2_list)}")
print(m2_list)
print(f"Points in Cluster 3 : {len(m3_list)}")
print(m3_list)

print(f"\nFinal value of m1 = {m1}")
print(f"Final value of m2 = {m2}")
print(f"Final value of m3 = {m3}")
```
